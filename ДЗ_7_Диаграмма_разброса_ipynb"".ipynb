```python
import pandas as pd
import gdown
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

def load_data(file_path):
    return pd.read_csv(file_path)
```

```python
def load_data(file_path):
    return pd.read_csv(file_path)

def analyze_age(df):
    df['age_analysis'] = df['age'].apply(lambda x: 'Старше 25' if x > 25 else 'Младше или равен 25')
    return df

def filter_by_sex_and_age(df):
    """Фильтрует DataFrame по условиям: пол — мужской, возраст — больше 30."""
    return df[(df['sex'] == 'male') & (df['age'] > 30)]

```

```python
# Задание 1 — загрузка данных и анализ возраста

file_id = '14PtmSWuSLCIzgqFHrq2SAE_FB1V6Ops4'

# Скачиваем файлы
file_path = 'web_clients_correct.csv'
gdown.download(f'https://drive.google.com/uc?id={file_id}', file_path, quiet=False)

data = load_data(file_path)
```

```python
# Задание 2 — фильтрация данных

print("\nЗадание 2: Фильтрация по полу и возрасту")
filter_by_sex_and_age(data)
```

```python
data.head()

plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='age', y='bill', color='blue', s=50)
plt.title("Диаграмма разброса зависимости возраста от расходов (Seaborn)")
plt.xlabel("Возраст")
plt.ylabel("Расходы (bill)")
plt.show()
```

```python
# Диаграмма разброса с помощью Plotly

fig = px.scatter(
    data,
    x='age',
    y='bill',
    title="Диаграмма разброса зависимости возраста от расходов (Plotly)",
    labels={'age': 'Возраст', 'bill': 'Расходы'},
    template="plotly_white"
)
fig.show()
```

```python
# Диаграмма регрессии (Seaborn)

plt.figure(figsize=(8, 6))
sns.regplot(data=data, x='age', y='bill', color='green', scatter_kws={'s': 50})
plt.title("Диаграмма регрессии зависимости возраста от расходов (Seaborn)")
plt.xlabel("Возраст")
plt.ylabel("Расходы (bill)")
plt.show()
```
